# 4. 시각적 분석

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

항공사 지연요인은 첫번째 구간이 12천, 두번째 구간이 15천이다.
ATC 지연요인은 첫번째구간이 2만으로 가장 높다
날씨 지연요인은 첫번째구간이 3만으로 가장 높다

```python
fig = plt.figure(figsize=(16,5))
ax1=fig.add_subplot(1,3,1)
ax2=fig.add_subplot(1,3,2)
ax3=fig.add_subplot(1,3,3)

ax1.hist(df['pct_carrier_delay'], bins=20)
ax2.hist(df['pct_atc_delay'],bins=20)
ax3.hist(df['pct_weather_delay'],bins=20)

ax1.set_title(['pct_carrier_delay'])
ax2.set_title(['pct_atc_delay'])
ax3.set_title(['pct_weather_delay'])
plt.show()
```

지연 유형별 중앙값, 사분위수, 이상치를 확인

1) 항공사 지연 : 20% 이상인 경우를 이상치로 판단, 그런데 100에 육박하는 극단적인 지연율도 있다.
2) ATC 지연: 항공사 지연보다 낮은 수준에서 이상치로 판단, 역시 100인 극단적인 지연도 있으나 항공사 지연요인보다 빈번하지  않다
3) 날씨 지연: 중앙값, IQR 자체가 작다.  이상치가 20이상인 경우이고 다른 요인들에 비해 상대적으로 빈번하진 않다.

```python
sns.boxplot(data=df)
```

```python
# 항공사별로 상자그림
```

항공사지연요인을 항공사별로 분석하면,

```python
sns.boxplot(x='airline',y='pct_atc_delay', data=df)
```

```python
sns.pairplot(df)
```

1. 항공사 요인 vs ATC 요인 간 상관계수가 0.14로 약한 상관관계가 있음을 확인할 수 있다.
2. 항공사 요인 vs 날씨 요인 간 상관계수가 0.16로 약한 상관관계가 있음을 확인할 수 있다.
   이것은 항공사 지연이 발생했을 때 기상 지연도 발생할 수 있다는 것을 의미하지만
   상당부분 두 지연요인은 독립적으로 발생하는 것으로 보인다.
4. ATC요인 vs 날씨 요인은 상관계수가 0.092로 거의 상관이 없다고 볼 수 있다.
   서로 독립적 으로 발생하는 경향이 있음을 의미한다.
